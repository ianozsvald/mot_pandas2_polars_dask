```python
import dask.dataframe as dd
import pandas as pd

from distributed import Client, progress, wait

import dask
import os
```

```python
dask.config.set(temporary_directory=os.getcwd())
client = Client()
client
```

```python
f = (
    dd
    .read_parquet('../../test_result.parquet',
                  dtype_backend='pyarrow',
                  pyarrow_options = {'filters':[('test_class_id', '==', 4)]},
                 )
    #.partitions[:3]
    .set_index('vehicle_id', shuffle='disk')
    .to_parquet('../../test_result_by_vehicle_id.parquet', overwrite=True, compute=False)
    .persist()
)

progress(f)
```

```python
f = (
    dd
    .read_parquet('../../test_result.parquet',
                  dtype_backend='pyarrow',
                  pyarrow_options = {'filters':[('test_class_id', '==', 4)]},
                 )
    #.partitions[:3]
    .set_index('first_use_date', shuffle='disk')
    .map_partitions(lambda x: x.sort_values(by=['vehicle_id', 'test_date']))
    .to_parquet('../../test_result_by_first_use_date.parquet', overwrite=True, compute=False)
    .persist()
)

progress(f)
```
